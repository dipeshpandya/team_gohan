```python
# Import required dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#reading csv file from link
df=pd.read_csv('https://docs.google.com/spreadsheets/d/1_ODNIn5n1k9RSVr_7gV2On4idZ6v0B2XPeMhCD18kio/pub?gid=1188863554&single=true&output=csv')
#df.head()
```

```python
chiefs = df['chief'].unique()
```

```python
x = 0
for chief in chiefs:
    print(chief)
    df.loc[df['chief'] == chief, 'chief_term'] = x
    #dfcoyp['chief_term'] = x
    x = x + 1
```

```python
dfwork=df[[
"decisionDirection",
"threeJudgeFdc",
"certReason",
"lcDisposition",
"partyWinning",
"majVotes",
"chief_term",
"decisionType",
"minVotes",
"caseDisposition"
]]
```

```python
X_dd = dfwork.drop(columns=["decisionDirection"])
Y_dd = dfwork[["decisionDirection"]]

X_dt = dfwork.drop(columns=["decisionType"])
Y_dt = dfwork[["decisionType"]]

X_pw = dfwork.drop(columns=["partyWinning"])
Y_pw = dfwork[["partyWinning"]]

X_cd = dfwork.drop(columns=["caseDisposition"])
Y_cd = dfwork[["caseDisposition"]]
```

```python
from sklearn.impute import SimpleImputer

# Impute missing values with the mean
dd_imputer = SimpleImputer(strategy='constant', fill_value=3)
dt_imputer = SimpleImputer(strategy='constant', fill_value=0)
pw_imputer = SimpleImputer(strategy='constant', fill_value=2)
cd_imputer = SimpleImputer(strategy='constant', fill_value=11)
```

```python
y_dd_imputed = pd.DataFrame(dd_imputer.fit_transform(Y_dd))
y_dt_imputed = pd.DataFrame(dt_imputer.fit_transform(Y_dt))
y_pw_imputed = pd.DataFrame(pw_imputer.fit_transform(Y_pw))
y_cd_imputed = pd.DataFrame(cd_imputer.fit_transform(Y_cd))
```

```python
# Impute missing values with the mean
dd_imputer = SimpleImputer(strategy='most_frequent')
dt_imputer = SimpleImputer(strategy='most_frequent')
pw_imputer = SimpleImputer(strategy='most_frequent')
cd_imputer = SimpleImputer(strategy='most_frequent')

X_dd_imputed = pd.DataFrame(dd_imputer.fit_transform(X_dd), columns=X_dd.columns)
X_dt_imputed = pd.DataFrame(dt_imputer.fit_transform(X_dt), columns=X_dt.columns)
X_pw_imputed = pd.DataFrame(pw_imputer.fit_transform(X_pw), columns=X_pw.columns)
X_cd_imputed = pd.DataFrame(cd_imputer.fit_transform(X_cd), columns=X_cd.columns)
```

```python
# Split the data into training and testing sets using random_state=42
X_dd_train, X_dd_test, y_dd_train, y_dd_test = train_test_split(X_dd_imputed, y_dd_imputed, random_state=42)
X_dt_train, X_dt_test, y_dt_train, y_dt_test = train_test_split(X_dt_imputed, y_dt_imputed, random_state=42)
X_pw_train, X_pw_test, y_pw_train, y_pw_test = train_test_split(X_pw_imputed, y_pw_imputed, random_state=42)
X_cd_train, X_cd_test, y_cd_train, y_cd_test = train_test_split(X_cd_imputed, y_cd_imputed, random_state=42)
```

```python
# Create the KNN model with 9 neighbors
knn_dd = KNeighborsClassifier(n_neighbors=10)
knn_dt = KNeighborsClassifier(n_neighbors=10)
knn_pw = KNeighborsClassifier(n_neighbors=10)
knn_cd = KNeighborsClassifier(n_neighbors=10)

# Fit the model to the training data
knn_dd.fit(X_dd_train, y_dd_train)
knn_dt.fit(X_dt_train, y_dt_train)
knn_pw.fit(X_pw_train, y_pw_train)
knn_cd.fit(X_cd_train, y_cd_train)
```

```python
train_dd_score = knn_dd.score(X_dd_train, y_dd_train)
test_dd_score = knn_dd.score(X_dd_test, y_dd_test)

train_dt_score = knn_dt.score(X_dt_train, y_dt_train)
test_dt_score = knn_dt.score(X_dt_test, y_dt_test)

train_pw_score = knn_pw.score(X_pw_train, y_pw_train)
test_pw_score = knn_pw.score(X_pw_test, y_pw_test)

train_cd_score = knn_cd.score(X_cd_train, y_cd_train)
test_cd_score = knn_cd.score(X_cd_test, y_cd_test)
```

```python
# Create the logistic regression classifier model with a random_state of 1
lr_model = LogisticRegression(max_iter=1000)

# Fit the model to the training data
lr_model.fit(X_dd_train, y_dd_train)

lr_predictions = lr_model.predict(X_dd_test)
# Calculate the accuracy score
acc_score = accuracy_score(y_dd_test, lr_predictions)

print(f"Training Data Score: {lr_model.score(X_dd_train, y_dd_train)}")
print(f"Testing Data Score: {lr_model.score(X_dd_test, y_dd_test)}")
print(f"Accuracy Score : {acc_score}")
```
